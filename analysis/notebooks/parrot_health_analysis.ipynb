```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ตั้งค่าให้แสดงกราฟใน notebook
%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("hpai-wild-birds-2022-2023.csv")
df.head()
```

```python
df.info()
df.columns
```

```python
top_species = df['Species'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_species.values, y=top_species.index, palette="crest")
plt.title("Top 10 Bird Species with HPAI Cases")
plt.xlabel("Number of Cases")
plt.ylabel("Species")
plt.tight_layout()
plt.show()
```

```python
top_province = df['Province'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_province.values, y=top_province.index, palette="flare")
plt.title("Top 10 Provinces with HPAI Cases")
plt.xlabel("Number of Cases")
plt.ylabel("Province")
plt.tight_layout()
plt.show()
```

```python
df['Confirmation date'] = pd.to_datetime(df['Confirmation date'], errors='coerce')
df['Month'] = df['Confirmation date'].dt.to_period('M')

monthly_cases = df['Month'].value_counts().sort_index()

plt.figure(figsize=(12, 4))
monthly_cases.plot(marker='o')
plt.title("Monthly HPAI Outbreaks")
plt.xlabel("Month")
plt.ylabel("Number of Cases")
plt.grid(True)
plt.tight_layout()
plt.show()
```
